```python
from sklearn.feature_extraction.text import CountVectorizer

documents = [
    "I love pizza",
    "Pizza is the best",
    "I love love pasta",
    "Pasta is great"
]

vectorizer = CountVectorizer()

X = vectorizer.fit_transform(documents)

print("Vocabulary: ", vectorizer.get_feature_names_out())
print("\nBoW Matrix:\n", X.toarray())
```
